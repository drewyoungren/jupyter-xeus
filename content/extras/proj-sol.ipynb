```python
def mag(v):
    # Calculate the magnitude 
    return sqrt(dot(v,v))
```

Use the NumPy function `array` to give a list (or tuple or other iterable) its vector-ness.

```python
u = array((1, -1, 3, 4, -2, -2))
```

```python
def proj(w, v):
    # vector projection of v onto w
    return dot(v, w) / dot(w, w) * w
```

```python
v = array((1, 1,1 ,1, 1, 1))
```

```python
proj(v, u)
```

```python
w = array((1, 1, 1, -1, -1, -1))
```

```python
x = array((-1, -1, 2, 2, -1, -1))
```

```python
plt.plot(v, label="$\\mathbf{v}$")
plt.plot(w, label="$\\mathbf{w}$")
plt.plot(x, label="$\\mathbf{x}$")
plt.legend();
```

These base vectors weren't chosen at random. They have a special property.

```python
dot(v,w), dot(w, x), dot(x, v)
```

They are *mutually orthogonal*.

```python
plt.plot(u, label="signal");
plt.plot(proj(v, u) + proj(w, u) + proj(x, u), label="proj");
plt.legend();
```
